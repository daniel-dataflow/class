```python
from unicodedata import category

import pandas as pd
```

```python
card_202007 = pd.read_csv("../card_data/csv/202007.csv")
print(card_202007)
```

```python
# 이용일, 금액, 대분류, 중분류, 소분류
card_202007 = card_202007[["이용일", "금액", "대분류", "중분류", "소분류"]]
print(card_202007)
```

```python
# 버스 22, 23, ... 뒤에 있는 숫자를 없애고 싶다!
# pandas 에서는 Method chaining을 deprecated 하려고 한다!
# .loc[??]= value
card_202007["중분류1"] = card_202007["중분류"].str.split(" ").str.get(0)
print(card_202007["중분류1"])
```

```python
card_202007["중분류1"]
```

```python
card_202007["중분류2"] = card_202007["중분류"].apply(lambda x: str(x).split(" ")[0])
print(card_202007)
```

```python
# 이용일 -> 날짜형식(타입)
card_202007.dtypes
```

```python
# 숫자를 바로 날짜로 바꾸면 안된다!!!
pd.to_datetime(card_202007["이용일"])
```

```python
# 숫자 -> 문자 -> 날짜
pd.to_datetime(card_202007["이용일"].astype(str))
card_202007.dtypes
```

```python
# 이뉴터블 하기 때문에 이용일에 덮어쓰기로 변경을 해야 제대로 변경된다.
card_202007["이용일"] =pd.to_datetime(card_202007["이용일"].astype(str))
```

```python
card_202007.dtypes
```

```python
# 년도, 월, 일, 요일 을 추가하고 싶다.
card_202007["년도"] = card_202007["이용일"].dt.year
card_202007["월"] = card_202007["이용일"].dt.month
card_202007["일"] = card_202007["이용일"].dt.day
card_202007["요일"] = card_202007["이용일"].dt.weekday
card_202007["요일_영어"] = card_202007["이용일"].dt.day_name()

```

```python
print(card_202007)
```

```python
weekday_list = ["월", "화", "수", "목", "금", "토", "일"]
card_202007["요일_한글"] = card_202007.apply(lambda x: weekday_list[x["요일"]], axis=1)
```

```python
print(card_202007)
```

```python
card = card_202007[["년도", "월", "일", "요일_한글", "금액", "대분류", "중분류1", "소분류"]]
print(card)
```
