```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
def cost(x, y, w):
    c = 0
    for i in range(len(x)):
        hx = w * x[i]
        loss = (hx - y[i]) ** 2
        c += loss

    return c / len(x)
```

```python
x = [1, 2, 3]
y = [1, 2, 3]

print(cost(x, y, -1))
print(cost(x, y, 0))
print(cost(x, y, 1))
print(cost(x, y, 2))
```

```python
for i in range(-30, 51):
    w = i / 10
    c = cost(x, y, w)
    plt.plot(w, c, "ro")

plt.title("cost function (loss function)")
plt.xlabel("weight")
plt.ylabel("cost")
plt.show()
```

```python
def gradient_descent(x, y, w):
    c = 0

    for i in range(len(x)):
        hx = w * x[i]
        loss = 2 * x[i] * (hx - y [i])

        """
        cost(w) = (w * x[i] - y[i])^2 의 미분
        cost(w) = w^2 * x[i]^2 -2 * w * x[i] * y[i]^2
        cost'(w) =  2 * w * x[i]^2  -2 *
        """

        c += loss

    return c / len(x)
```

```python
w = 10 
old = 0
for i in range(1000):
    c = cost(x, y, w)
    grad = gradient_descent(x, y, w)
    # 0.1 = learning rate
    w -= 0.1 * grad

    print(f"[{i:03d}] cost: {c} \t old: {old} \t weight: {w}") 

    old = c
print(f"weight : {w}  train : {i+1} 회")
# 
```

```python
def fit(x, y):
    w, old = 10, 100 

    # 학습할때 변화율 -> 비슷한 값이 나올때까지 변경을 시켜서 값을 추론한다.
    learning_rate = 0.01

    for i in range(1000):
        c = cost(x, y, w)
        grad = gradient_descent(x, y, w)
        w -= learning_rate * grad

        print(f"[{i:03d}] cost: {c} \t old: {old} \t weight: {w}") 

        old = c
    
    return w
```

```python
def predict(x, w):
    hx = w * np.array(x)
    return list(hx)
```

```python
# 학습시키기
x_train =[1, 2, 3, 4, 5]
y_train = [4, 7, 10, 13, 16]
w = fit(x_train, y_train)
print(w)
```

```python
x_test = [6, 7, 8, 9, 10]
y_test = [19, 22, 25, 28, 31]
y_pred = predict(x_test, w)
print(y_pred)
```

```python
class MyLinearRegression:
    def __init__(self, weight=10, learning_rate=0.01):
        self.w = weight
        self.lr = learning_rate

# 값을 비교
    def cost(self, x, y):
        c = 0
        for i in range(len(x)):
            hx = self.w * x[i]
            loss = (hx - y[i]) ** 2
            c += loss

        return c / len(x)
    
    def gradient_descent(self, x, y):
        c = 0

        for i in range(len(x)):
            hx = self.w * x[i]
            loss = 2 * x[i] * (hx - y [i])

            c += loss

        return c / len(x)
    
    def fit(self, x, y):
        old = 100 

        for i in range(1000):
            c = self.cost(x, y)
            grad = self.gradient_descent(x, y)
            self.w -= self.lr * grad

            print(f"[{i:03d}] cost: {c} \t old: {old} \t weight: {self.w}") 

            old = c
        
    def predict(self, x):
        hx = self.w * np.array(x)
        return list(hx)
```

```python
x_train = [1, 2, 3, 4, 5]
y_train = [2, 4, 6, 8, 10]

linear_regression = MyLinearRegression(learning_rate=0.01)
linear_regression.fit(x_train, y_train)
print(f"weight : {linear_regression.w}")
# 발산하고 있다. => 원하는 값에서 더욱더 멀어지고 있다
```

```python
x_test =[10, 20, 30]
y_pred = linear_regression.predict(x_test)
print(y_pred) 
```
