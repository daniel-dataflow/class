```python
# 1. 데이터셋 준비
x = tips[["total_bill"]]
y = tips[["tip"]]

print(x)
print(y)
```

```python
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LinearRegression
```

```python
# 2. 데이터 분할 * test_size : 나눌때 비율 : 랜덤으로 데이터를 30%로 남겨 놓겠다 / random_state 시드를 고정해줘 - 지정값으로 고정해 놓는다는 의미로
x_train, x_test, y_train, y_test = train_test_split(x, y, test_size=0.3, random_state=1)
```

```python
# 3. 모델 선택 - 숫자로 되어 있기 때문에 선형회기분석을 사용한다.
model = LinearRegression()
```

```python
# 4. 모델 학습
model.fit(x_train, y_train)
```

```python
# 5. 예측 및 평가
predict = model.predict(x_test)
# predict y의 예측값으로 보인다.
```

```python
# 6 검증
model.score(x_test, y_test)
```

```python
# 총 요금이 30달라인 경우에, 예상 팁은?
model.predict([[30]])
```
